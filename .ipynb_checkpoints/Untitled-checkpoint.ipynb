```python
from csv import reader

file = open('status.txt', encoding="utf8")
read_file = reader(file)
rivit = list(read_file)
```

```python
def explore_data(dataset, start, end):
    for row in dataset:
        print(row)
```

```python
#explore_data(rivit, 0, len(rivit))
```

```python
siivotut = []
for row in rivit:
    if "Package:" in str(row) or "Description:" in str(row) or "Depends:" in str(row):
        siivotut.append(str(row))
        
```

```python
#explore_data(siivotut, 0, len(siivotut))
```

```python
tuplet = []
while len(siivotut) > 0:
    if "Package:" in siivotut[0]:
        tempList = []
        tempList.append(siivotut.pop(0))
        while (True):
            if len(siivotut) <= 0:
                break
            if "Package:" in siivotut[0]:
                tuplet.append(tuple(tempList))
                break
            else:
                tempList.append(siivotut.pop(0))
    
```

```python
for row in tuplet:
    print(row)
```

```python
f = open("index.html","w+")
for row in tuplet:
    packageName = row[0]
    packageName.split(":")
    print(packageName)
    #f.write("{} \n".format(packageName[1]))
    
f.close()
```
